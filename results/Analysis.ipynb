```python
import numpy as np

real_real = 'real-real.txt'
real_decoy = 'real-decoy.txt'

with open(real_real, 'r') as f:
    rr = f.readlines()

with open(real_decoy, 'r') as f:
    rd = f.readlines()

rr_x = [i for i in range(len(rr))]
rr_y = [float(i) for i in rr]

rd_x = [i for i in range(len(rd))]
rd_y = [float(i) for i in rd]
```

```python
import matplotlib.pyplot as plt

plt.scatter(rr_x,rr_y, label='real', color='blue', s=25, marker="o")
plt.scatter(rd_x,rd_y, label='decoy', color='green', s=25, marker="o")

plt.legend()
plt.savefig('plot.png')
plt.show()
```

```python
print(np.median(rr_y))
print(np.median(rd_y))

print(np.std(rr_y))
print(np.std(rd_y))
```
